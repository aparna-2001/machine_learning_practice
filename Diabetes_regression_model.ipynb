```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_diabetes
```

```python

data = load_diabetes()

diabetes_data = pd.DataFrame(data.data, columns=data.feature_names)
diabetes_data['target'] = data.target
diabetes_data.head()
```



*   The data contains 442 rows and 10 columns.

*   All the data is in float.

* All columns contain 442 values, so there are no null values.





*   Modeling regression between blood pressure and target (diseases progression score after one year)

*   Independent variable, x = blood pressure

*   Dependent variable, y = target





```python
#adding constant to x values, so the y line not pass through the origin

x = sm.add_constant(diabetes_data['bp'])
y = diabetes_data['target']
x.head()
```

```python
#training and testing data

train_x, test_x, train_y, test_y = train_test_split(x,
                                                    y,
                                                    train_size = 0.8,
                                                    random_state = 100)
```

```python
#fitting the model

diabetes_model = sm.OLS(train_y, train_x).fit()
print(diabetes_model.params)
```



*   Estimated model can be written as in the form of straight line;

    y = mx + c ;

    target = 675.853574*bp + 153.041444
